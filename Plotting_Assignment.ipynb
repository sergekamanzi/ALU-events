```python
# YOUR CODE HERE
# Histogram for CO2 emissions with adjusted bin size
plt.figure(figsize=(10,6))
sns.histplot(df['co2'], bins=30, kde=True)  # Adjust bins size here, try different values like 10, 20, 30, 50
plt.title('CO2 Emissions Distribution')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE
# Histogram for fuel efficiency
plt.figure(figsize=(10,6))
sns.histplot(df['comb'], bins=30, kde=True)  # Assuming 'comb' is combined fuel efficiency, adjust if necessary
plt.title('Fuel Efficiency Distribution')
plt.xlabel('Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10,6))
sns.histplot(df['displ'], bins=30, kde=True)  # Assuming 'displ' stands for engine displacement
plt.title('Engine Displacement Distribution')
plt.xlabel('Displacement (L)')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Create a heatmap for correlation between numeric features
plt.figure(figsize=(12,8))
corr_matrix = numeric_df.corr()  # Compute the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm_r', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
